# Imports

````python
import numpy as np
import tensorflow as tf
from tensorflow.compat.v1.saved_model import load, tag_constants
````

# Constants

````python
X_TEST_PATH = 'MNIST/load_xtest.csv'
Y_TEST_PATH = 'MNIST/load_ytest.csv'
````

# Load the Data

````python
y_test = np.loadtxt(Y_TEST_PATH, delimiter=',', dtype=int)
print(f'Shape of y_test {y_test.shape}')
````

````python
x_test = np.loadtxt(X_TEST_PATH, delimiter=',', dtype=float)
print(f'Shape of x_test {x_test.shape}')
````

# Create Session

````python
graph = tf.Graph()
sess = tf.compat.v1.Session(graph=graph)
````

# Load the Model

```
tf.compat.v1.saved_model.load(sess,
                           [tag_constants.SERVING],
                           EXPORT_DIR)
```

````python
load(sess=sess, tags=[tag_constants.SERVING], export_dir='SavedModel')
````

# Run Session & Make Prediction

````python
X = graph.get_tensor_by_name('X:0') # here we are  getting our x placeholder which held onto our x values
#and we add the :0 at the end the get the actual result of an operation and not just the operation
#itself. The 0 is an index if the operation resulted in 2 values => 0=first result => 1=2nd result

y_pred = graph.get_tensor_by_name('accuracy_calc/prediction:0')
````

````python
prediction = sess.run(fetches=y_pred, feed_dict={X:x_test})
````

# Check Results

````python
y_test[:5]
````

````python
prediction[:5]
````

````python
correct_pred = (y_test == prediction).sum()
correct_per = correct_pred / y_test.shape[0]
print(f"The percentage of the test dataset we predicted correctly is {correct_per:0.2%}")
````

# Cleanup

````python
sess.close()
tf.compat.v1.reset_default_graph()
````
